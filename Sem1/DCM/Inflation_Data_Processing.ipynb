```python
import random
import pandas as pd
```

```python
# Load the CSV file containing US inflation rates
df = pd.read_csv("./data/raw/US_inflation_rates.csv")
df["date"] = df["date"].str.slice(0,4)
df[:25]
```

```python
# Group by year and calculate the mean inflation rate for each year
df_yearly = df.groupby("date").mean().reset_index()
df_yearly.head()
```

```python
#check if missing years
df_yearly["date"].tolist() != list(str(x) for x in range(1947, 2026))
```

```python
# Add a column indicating whether each year was a recession year with different formats
recession_years = [1949, 1953, 1957, 1960, 1970, 1974, 1980, 1981, 1990, 2001, 2008, 2009, 2020]

positive_formats = ["Yes", "True", "TRUE", "1", 1, True]
negative_formats = ["No", "False", "FALSE", "0", 0, False]

def get_recession_status_random(year):
    if year in recession_years:
        return random.choice(positive_formats)
    else:
        return random.choice(negative_formats)
    
df_yearly['Is_Recession'] = df_yearly['date'].astype(str).str[:4].astype(int).apply(get_recession_status_random)
df_yearly[:10]
```

```python
df_yearly.to_csv("./data/ready/US_inflation_rates_ready.csv", index=False)
```
