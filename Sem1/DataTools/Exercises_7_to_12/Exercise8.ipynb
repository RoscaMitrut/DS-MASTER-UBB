```python
import hashlib
```

```python
def get_checksum(file_path):
    with open(file_path, "rb") as f:
        file_bytes = f.read()
        return hashlib.sha256(file_bytes).hexdigest()
```

```python
filename = "temp.txt"
text = "Lorem ipsum dolor sit amet, consectetuer adipiscing elit. Aenean commodo ligula eget dolor. Aenean massa. Cum sociis natoque penatibus et magnis dis parturient montes, nascetur ridiculus mus."
```

```python
with open(filename, "w") as f:
    f.write(text)
    
checksum1 = get_checksum(filename)
print(f"Original Checksum: {checksum1}")
```

```python
with open(filename, "w") as f:
    f.write(text + "A")

checksum2 = get_checksum(filename)
print(f"Modified Checksum: {checksum2}")
```

```python
if checksum1 != checksum2:
    print("STATUS: Checksum changed. File modification detected.")
```

## Questions

- Did the checksum change?<br>
Yes. Adding a single character resulted in a completely different hash string.

- Why is it useful in collaborative research?<br>
Datasets can be verified to ensure they were not corrupted during download or tampered with by someone else. If the hashes match, the files are identical.
