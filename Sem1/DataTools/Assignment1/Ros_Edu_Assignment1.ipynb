```python
import pandas as pd

df = pd.read_excel("./Teams_R3dw4rd_/data_analysis_raw.xlsx")
df.head()
```

```python
# Region is not available, so i will calculate revenue of each supplier

df_revenue = df.drop(labels=["Year", "Month"], axis=1).groupby(by="Supplier",dropna=True).sum(numeric_only=True).rename(columns={"Profit (Month)":"Total Profit"}).reset_index()
df_revenue.head()
```

```python
df_top3 = df.drop(labels=["Year", "Month","Supplier"], axis=1).groupby(by=["Category","Brand"],dropna=True).sum(numeric_only=True).rename(columns={"Profit (Month)":"Profit"}).sort_values(by="Profit").reset_index()
df_top3[:3]
```

```python
df_monthly_profit = df.drop(labels=["Supplier", "Category", "Brand"], axis=1).groupby(by=["Year", "Month"]).sum(numeric_only=True).sort_values(by=["Year","Month"]).rename(columns={"Profit (Month)":"Profit"}).reset_index()
df_monthly_profit["Growth %"] = df_monthly_profit["Profit"].pct_change()*100
df_monthly_profit
```

```python
df_monthly_profit.to_excel("temp.xlsx", index=False)
```
