# Simple problems

```python
def is_ten(a,b):
    return (a==10) or (b==10) or (a+b==10)

def sum_of_first_n(n):
    sum = 0
    for i in range(n+1):
        sum+=i
    return sum

def sum_of_first_even_n(n):
    sum = 0
    for i in range(0,n*2+1,2):
        sum += i
    return sum

def is_perfect_square(n):
    import math
    a = int(math.sqrt(n))
    return a*a==n

def is_prime(n):
    if n<2:
        return False
    for i in range(2,n-1):
        if n%i==0:
            return False
    return True

def find_primes_up_to_n(n):
    primes = []
    for i in range(2,n):
        if is_prime(i):
            primes.append(i)
    return primes

def find_largest_prime_smaller_than_n(n):
    for i in range(n,2,-1):
        if is_prime(i):
            return i

def sum_of_primes_in_range(n,m):
    sum=0
    for i in range(n,m):
        if is_prime(i):
            sum+=i
    return sum
```

```python
print("pb1: ",is_ten(7,3))
print("pb2: ",sum_of_first_n(5))
print("pb3: ",sum_of_first_even_n(3))
print("pb4: ",is_perfect_square(16))
print("pb5: ",is_prime(29))
print("pb6: ",find_primes_up_to_n(20))
print("pb7: ",find_largest_prime_smaller_than_n(15))
print("pb8: ",sum_of_primes_in_range(10,20))
```

# Functions

```python
def create_list():
    """
    Reads a list of integers from the user.
    Returns:
        list[int]: The list of integers entered by the user.
    """
    numbers = []
    n = int(input("Nr. of integers: "))
    for i in range(n):
        nr = int(input(f"Enter integer {i+1}: "))
        numbers.append(nr)
    return numbers

def filter_even_integers(lst):
    """
    Returns a list of even integers from the given list.
    Args:
        lst (list[int]): Input list.
    Returns:
        list[int]: List of even numbers.
    """
    res = []
    for el in lst:
        if el % 2 == 0:
            res.append(el)
    return res

def sum_excluding_twelve(lst):
    """
    Computes the sum of numbers in the list, excluding the number 12.
    Args:
        lst (list[int]): Input list.
    Returns:
        int: Sum of all numbers except 12.
    """
    total = 0
    for el in lst:
        total += el
    total -= lst.count(12) * 12
    return total

def find_min_max(lst):
    """
    Finds the minimum and maximum values in the list.
    Args:
        lst (list[int]): Input list.
    Returns:
        tuple: (minimum value, maximum value)
    """
    min_val = min(lst)
    max_val = max(lst)
    return min_val, max_val

def convert_list_to_set(lst):
    """
    Converts a list into a set (removing duplicates).
    Args:
        lst (list[int]): Input list.
    Returns:
        set[int]: Set of unique elements.
    """
    return set(lst)

def check_friendship(a, b):
    """
    Checks whether two numbers are 'friends' (amicable numbers).
    Two numbers are friends if each equals the sum of the divisors of the other.
    Args:
        a (int): First number.
        b (int): Second number.
    Returns:
        bool: True if they are friends, False otherwise.
    """
    def sum_of_div(n):
        total = 0
        for i in range(1, n):
            if n % i == 0:
                total += i
        return total

    return sum_of_div(a) == b and sum_of_div(b) == a

def generate_grouped_sequence(n):
    """
    Generates a sequence: 1, 2, 2, 3, 3, 3, 4, 4, 4, 4, ... up to the given length n.
    Args:
        n (int): The desired sequence length (number of groups, not total elements).
    Returns:
        str: The sequence as a comma-separated string.
    """
    string = ""
    for i in range(1, n + 1):
        string += (str(i) + ",") * i
    return string[:-1]
```

```python
lst = create_list()
print("List entered:", lst)

evens = filter_even_integers(lst)
print("Even numbers:", evens)
print("Number of even numbers:", len(evens))

total_sum = sum_excluding_twelve(lst)
print("Sum excluding 12:", total_sum)

min_val, max_val = find_min_max(lst)
print(f"Minimum: {min_val}, Maximum: {max_val}")

s = convert_list_to_set(lst)
print("Set version of list:", s)

a = int(input("\nEnter first number to check for friendship: "))
b = int(input("Enter second number: "))
if check_friendship(a, b):
    print(f"{a} and {b} are friends!")
else:
    print(f"{a} and {b} are NOT friends.")

n = int(input("\nEnter how many terms to generate in the sequence: "))
seq = generate_grouped_sequence(n)
print("Generated sequence:", seq)
```

# Lists

```python
def create_list():
    n = int(input("Enter number of elements: "))
    my_list = []
    for i in range(n):
        x = int(input(f"Element {i + 1}: "))
        if x < 0:
            print("Only natural numbers are allowed. Try again.")
            return create_list()
        my_list.append(x)
    return my_list

def list_to_string(lst):
    return " ".join(str(x) for x in lst)

def is_power_of_two(n):
    return n > 0 and (n & (n - 1)) == 0

def sum_powers_of_two(lst):
    return sum(x for x in lst if is_power_of_two(x))

def filter_not_powers_of_two(lst):
    return [x for x in lst if not is_power_of_two(x)]

def delete_sequence_between_evens(lst):
    first_even = None
    last_even = None
    for i in range(len(lst)):
        if lst[i] % 2 == 0:
            if first_even is None:
                first_even = i
            last_even = i
    if first_even is not None and last_even is not None:
        return lst[:first_even] + lst[last_even + 1:]
    return lst[:]

def delete_all_evens(lst):
    return [x for x in lst if x % 2 != 0]

def shift_longest_even_sequence(lst):
    max_len = 0
    max_start = -1

    current_len = 0
    current_start = -1

    for i in range(len(lst)):
        if lst[i] % 2 == 0:
            if current_len == 0:
                current_start = i
            current_len += 1
            if current_len > max_len:
                max_len = current_len
                max_start = current_start
        else:
            current_len = 0

    if max_len == 0:
        return lst[:]

    longest_seq = lst[max_start:max_start + max_len]
    remaining = lst[:max_start] + lst[max_start + max_len:]
    return longest_seq + remaining

def print_menu():
    print("\n=== MENU ===")
    print("1. Create list of natural numbers")
    print("2. Convert list to string")
    print("3. Sum of elements that are powers of 2")
    print("4. Filter elements that are NOT powers of 2")
    print("5. Delete sequence between first and last even numbers")
    print("6. Delete all even numbers")
    print("7. Shift to beginning the longest sequence of even numbers")
    print("0. Exit")


def main():
    my_list = []

    while True:
        print_menu()
        choice = input("Choose an option: ")

        if choice == '1':
            my_list = create_list()
            print("List created:", my_list)
        elif choice == '2':
            print("List as string:", list_to_string(my_list))
        elif choice == '3':
            print("Sum of powers of 2:", sum_powers_of_two(my_list))
        elif choice == '4':
            print("Filtered list (not powers of 2):", filter_not_powers_of_two(my_list))
        elif choice == '5':
            my_list = delete_sequence_between_evens(my_list)
            print("Updated list:", my_list)
        elif choice == '6':
            my_list = delete_all_evens(my_list)
            print("Updated list:", my_list)
        elif choice == '7':
            my_list = shift_longest_even_sequence(my_list)
            print("Updated list:", my_list)
        elif choice == '0':
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Try again.")

main()
```
